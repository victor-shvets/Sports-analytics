# Libs

```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
from pandasql import sqldf
```

# Def

```python
def api(param):
    
    url = 'https://api-football-v1.p.rapidapi.com/v3/' + param

    headers = {
        "X-RapidAPI-Key": "e613e8d615msha96b42823ece3a1p12cd95jsn12a8f361c699",
        "X-RapidAPI-Host": "api-football-v1.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers)
    
    return response.json()
```

# Load data

## Season

```python
# api('seasons)

season_id = 2021
```

## Leagues

Run api

```python
jLeagues = api('leagues')
```

Transform into dataframe

```python
# Normalize from json
dfLeagues = pd.json_normalize(jLeagues['response'])

# Rename columns
dfLeagues.columns = dfLeagues.columns.str.replace('.', '_')

# Save into file
dfLeagues.drop('seasons', axis=1).to_csv('leagues.csv',index=False)

# Get league id
league_id = dfLeagues[(dfLeagues['country_name'] == 'England')\
                      & (dfLeagues['league_name'] == 'Premier League')]['league_id'].values[0]
```

## Teams

Run api

```python
jTeams = api(f'teams?league={league_id}&season={season_id}')
```

Transform into dataframe

```python
# Normalize from json
dfTeams = pd.json_normalize(jTeams['response'])

# Rename columns
dfTeams.columns = dfTeams.columns.str.replace('.', '_')

# Save into file
dfTeams.to_csv('teams.csv',index=False)
```
